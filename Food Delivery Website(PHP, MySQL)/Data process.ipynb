```python
def det(i):
    r = []
    n = list(detail[detail['restaurant_id']==i]['name'])
    p = list(detail[detail['restaurant_id']==i]['price'])
    num = random.sample(range(len(detail[detail['restaurant_id']==i]['name'])),3)
    amount = [random.randint(1,2),random.randint(1,2),random.randint(1,2)]
    r.append(n[num[0]]+' * '+str(amount[0]))
#     r.append(str(p[num[0]]))
    r.append(',')
    r.append(n[num[1]]+' * '+str(amount[1]))
#     r.append(str(p[num[1]]))
    r.append(',')
    r.append(n[num[2]]+' * '+str(amount[2])+'",'+str(p[num[0]]*amount[0]+p[num[1]]*amount[1]+p[num[2]]*amount[2]))
#     r.append(str(p[num[2]]))
    return ''.join(r)
det(1)
```
